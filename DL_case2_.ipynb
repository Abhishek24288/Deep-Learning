```python
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
# yaa han pay CIFAR-10 dataset import kar rahe hain (images dataset with 10 classes)
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
#Training + testing data load key hain yaa han pay
X_train = X_train / 255.0
X_test = X_test / 255.0
# yaa han pay Image normalization (0вЂ“255 в†’ 0вЂ“1)
y_train = to_categorical(y_train,10)
y_test = to_categorical(y_test,10)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
#Sequential model (layer by layer)
model = Sequential([
    Conv2D(32,(3,3),activation='relu',padding='same',input_shape=(32,32,3)),
    Conv2D(32,(3,3),activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(64,(3,3),activation='relu',padding='same'),
    Conv2D(64,(3,3),activation='relu'),
    MaxPooling2D(2,2),
    # 2D в†’ 1D convert
    Flatten(),
    Dense(256,activation='relu'),
    Dense(10,activation='softmax')
])
```

```python
#Model ready for training
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_train,y_train,epochs=10,validation_data=(X_test,y_test))
```
